## Preguntas

Por año cuantos libros ha publicado un autor, recuerde que los años no trabajados deberia de aparecer en cero. considere como primer registro por autor el primer año en que publico y el ultimo el año maximo encontrado.

Ejemplo, si Uayeb public en 2009 dos libros, y en 2011 tres. y esas son las dos unicas entradas para este autor. la salida para el deberia de ser:

```python
pd.DataFrame({ "Autor" : ["Uayeb","Uayeb","Uayeb"] , "Year" : [2009,2010,2011]  , "Conteo" : [2,0,3] })
```

## Solucion Ejercicio 1

```python
e1=amazon.sort_values(['Author','Year'], ascending=True)
```

```python
e1_new1=e1.groupby(['Author','Year']).agg({'Author':['count']})
```

```python
e1_new1=e1_new1.reset_index()
```

```python
e1_new1.columns=e1_new1.columns.droplevel(1)
```

```python
e1_new1.columns=['Author','Year','Conteo']
```

```python
e1Autores=amazon.loc[:, ['Author']]
e1anos=amazon.loc[:, ['Year']]
```

```python
e1Autores=e1Autores.drop_duplicates()
```

```python
e1anos=e1anos.drop_duplicates()
```

```python
e1_new2=e1Autores.merge(e1anos,how='cross')
```

```python
e1_new2=e1_new2.sort_values(['Author','Year'], ascending=True)
```

```python
e1_solucion=e1_new1.merge(e1_new2, how='right', on=['Author','Year'])
```

```python
e1_solucion=e1_solucion.fillna(0)
```

```python
e1_solucion
```

```python
e3=amazon.sort_values(['Genre','Year'], ascending=[True,True])
```

## Solucion Ejercicio #4

```python
e4=amazon.loc[:,['Author','Name','Year','User Rating']]
```

```python
e4_min=e4.sort_values(['Author','Year'], ascending=[True,True])
e4_min=e4_min.groupby(['Author','Year'], as_index=False).agg({'User Rating':['mean']})
e4_min.columns=e4_min.columns.droplevel(1)
e4_min['Row_Min']=e4_min.groupby(['Author'], as_index=False).cumcount()
```

```python
e4_max=e4.sort_values(['Author','Year'], ascending=[True,False])
e4_max=e4_max.groupby(['Author','Year'], as_index=False).agg({'User Rating':['mean']})
e4_max=e4_max.sort_values(['Author','Year'], ascending=[True,False])
e4_max.columns=e4_max.columns.droplevel(1)
e4_max['Row_Max']=e4_max.groupby(['Author'], as_index=False).cumcount()
```

```python
e4_max1=e4_max.loc[e4_max['Row_Max']==0,['Author','User Rating']]
e4_min1=e4_min.loc[e4_min['Row_Min']==0,['Author','User Rating']]
```

```python
e4_max1.columns=['Author','max_rate']
e4_min1.columns=['Author','min_rate']
```

```python
e4_new=e4_max1.merge(e4_min1, how='inner', on=['Author'])
```

```python
e4_new= e4_new.loc[e4_new['max_rate']< e4_new['min_rate'],['Author']]
```

```python
e4_new
```

## Solucion Ejercicio #5

```python
e5= amazon.loc[:,['Year','Name','Author','User Rating']]
```

```python
e5=e5.sort_values(['Year','User Rating'], ascending=True)
```

```python
percentil=pd.DataFrame()
```

```python
percentil['Percentil']=amazon.groupby('Year')['User Rating'].quantile(0.7)
```

```python
percentil=percentil.reset_index()
```

```python
e5_new1=e5.merge(percentil,how='left', on='Year')
```

```python
e5_new1=e5_new1.loc[e5_new1['User Rating']> e5_new1['Percentil'],:]
```
